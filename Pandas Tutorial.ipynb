```python
#Fill the null value 
#fill null value with specific value
df = pd.read_csv('C:/Users/proth/Desktop/Github/NumPy/Car_data.csv', sep= ';')
df.fillna(0).head()
```

```python
df.fillna(4).head()
```

```python
#Fill null value by column name as dictionary
df.fillna({'Car_Name': 'None', 'Selling_Price': 5}).head()
```

```python
#fill value with method use 'ffill' as copy value from previous value
df.fillna(method='ffill').head()
```

```python
#fill value with method use 'bfill' /'pad' as copy value from after value
df.fillna(method='bfill').head()
```

```python
# fill value with specific value and use limit for how many values
df.fillna(0,limit=1).head()
```

```python
df.fillna(1,limit=2).head()
```

```python
df.fillna(method='ffill',limit=1).head()
```

# Pandas replace() function

```python
#Replace value with another value also work for integer value
df.replace(to_replace='Manual', value='Automated').head()
```

```python
df.replace('Manual', 'Automated').head()
```

```python
#replace multiple value with multiple value
df.replace(['Yes','no'], [1,0]).head()
```

```python
df.replace({'Car_Name':'ritz'}, 'AUDI').head()
```

```python
df.replace({'Car_Name':'ritz', 'Fuel_Type':'Petrol'}, 'AUDI').head()
```

```python
#replace all total string value
df.replace('[A-Za-z]',0, regex=True).head()
```

```python
df.replace({'Car_Name':'[A-Za-z]' }, 0, regex=True).head()
```
